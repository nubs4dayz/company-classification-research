#Description

Our model uses a pipeline that employs zero-shot classification with the 'valhalla/distilbart-mnli-12-3' model. Rather than dividing the dataset into training and testing sets, we use all descriptions from the dataset as input. To prevent any potential disruption to the model, we also change the names of the sectors. These names are obtained using TF-IDF vectorization and choosing the most suitable names for each sector. 

Despite these measures, the sectors that previously showed the lowest F1 scores (Real Estate, Consumer Staples, Consumer Discretionary, and Industrials) are still showing low F1 scores again, indicating high levels of noise in these sectors.

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model='valhalla/distilbart-mnli-12-3', device=0, batch_size=8)
```

#Predictions

```python
pred = classifier(list(data["busdesc"]), labels)
```

#Results

```python
result = []
for item in pred:
  result.append(item["labels"][0])
```

```python
import math

true = []
for item in data["gind"]:
  key = math.floor(item/10000)
  true.append(sector[key])
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score

print(confusion_matrix(true, result))
print(classification_report(true, result))
print("F1 score is: "+ (str)(f1_score(true, result, average='micro')))
```
